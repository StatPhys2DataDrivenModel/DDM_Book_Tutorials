```python
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.special
%matplotlib inline

```

```python
# GERMAN TANK PROBLEM
M=10
yM=100
norm=scipy.special.binom(yM,M)*(M-1)/yM
avN=(M-1)/(M-2)*(yM-1)
dist=np.zeros((251,2));
for N in range(yM+1):
    dist[N,0]=N
    dist[N,1]=0
for i in range(yM+1,251):
    N=i-1
    dist[i,0]=N
    dist[i,1]=norm/scipy.special.binom(N,M)

# Plot the Posterior Distribution
plt.plot(dist[:,0],dist[:,1],color='black')
plt.axvline(x=avN,color='black',linestyle="--")
plt.axvline(x=yM,color='black',linestyle="--")
plt.legend()
plt.xlabel('$N$')
plt.ylabel('Posterior distribution $P(N|M,y_M)$')
plt.ylim(0,0.1)
plt.xlim(60,200)
plt.text(112,0.06,r'$\langle N \rangle$')
plt.text(93,0.06,r'$y_M$')
plt.text(160,0.002,r'$\sim N^{-M}$')
plt.savefig('germantank.pdf')
plt.show()
```

```python
# LAPLACE PROBLEM

dist=np.zeros((99,2));
thstar=0.490291
th=0
for i in range(0,99):
    th=th+0.01
    dist[i,0]=th
    dist[i,1]=thstar*np.log(th)+(1-thstar)*np.log(1-th)

# Plot the Posterior Distribution
plt.plot(dist[:,0],dist[:,1],color='black')
plt.axvline(x=thstar,color='black',linestyle="--")
plt.legend()
plt.xlabel(r'$\theta$')
plt.ylabel(r'$f_{\theta^*}(\theta)$')
plt.ylim(-2.2,-0.6)
plt.xlim(0,1)
plt.text(0.5,-1.5,r'$\theta^*$')
plt.savefig('laplace.pdf')
plt.show()
```
