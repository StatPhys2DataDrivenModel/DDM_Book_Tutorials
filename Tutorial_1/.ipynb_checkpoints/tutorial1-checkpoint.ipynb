```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["font.family"] = "serif"
plt.rcParams["mathtext.fontset"] = "dejavuserif"
plt.rcParams["figure.figsize"] = (10, 8)
plt.rcParams["font.size"] = 26
```

```python

#read time series t, x(t),y(t) generated with a drift D 
data=np.loadtxt('Data/dataM1000d2.5.dat')

#read the number of data point from the data
N=np.size(data[:,1])

#compute one estimate of D
r=np.sqrt(np.power(data[:,1],2)+np.power(data[:,2],2))
D0=np.power(r[N-1],2)/data[N-1,0]/4
print("First estimate of D:",D0)

#compute another estimate of D
diffdata=np.diff(data,axis=0)
dr2=(np.power(diffdata[:,1],2)+np.power(diffdata[:,2],2))/diffdata[:,0]
D1=np.mean(dr2)/4.
print("Second estimate of D:",D1)


# Plot the trajectory  
plt.plot(data[:,1],data[:,2],color='black')
plt.xlabel('$x$ ($\mu$m)')
plt.ylabel('$y$ ($\mu$m)')
#plt.title('Trajectory data M=1000')
plt.text(-8, 25, 'A', fontweight='bold')
plt.tight_layout()
plt.savefig('trajectory_dataM1000d2_5.pdf')
plt.show()
#Plot separately the x and y and the displacement  r=sqrt(x^2+y^2) in function of the time  and
#plt.plot(data[:,0],data[:,1])
#plt.show()
#plt.plot(data[:,0],data[:,2])
#plt.show()
plt.plot(data[:,0],r,color='black')
plt.plot(data[:,0],np.sqrt(4*D1*data[:,0]),color='black',linestyle='-.',label='${\cal D}$=2.47 $\mu$m$^2$s$^{-1}$')
plt.plot(data[:,0],np.sqrt(4*D0*data[:,0]),color='black',linestyle='--',label='${\cal D}$=1.20 $\mu$m$^2$s$^{-1}$')
plt.xlabel('$t$ (s)')
plt.ylabel('$r$ ($\mu$m)')
plt.legend()
plt.text(-10, 66, 'B', fontweight='bold')
#plt.title('Trajectory data M=1000')
plt.tight_layout()
plt.savefig('rvst_dataM1000d2_5.pdf')
plt.show()
```

```python
def posterior(namefile):
    data=np.loadtxt(namefile)
    #read the number of data point from the data
    N=np.size(data[:,1])
    diffdata=np.diff(data,axis=0)
    dr2=(np.power(diffdata[:,1],2)+np.power(diffdata[:,2],2))/diffdata[:,0]/4
    B=np.sum(dr2)
    #calculate the most probable value of the diffusion coefficient
    d0=B/(N-1)
    #calculate the average value of the diffusion coefficient
    d1=B/(N-3)
    #calculate the standard deviation of the diffusion coefficient
    d2=B/((N-3)*math.sqrt(N-4))
    print(N,d0,d1,d2)
    #compute posterior distribution
    dist=np.zeros((400,2));
    d=0
    for i in range(400):
        d=d+0.01
        dist[i,0]=d
        #dist[i,1]=np.exp(-b/d)*np.power(d,-N+1)*np.power(b,N-2)/(math.factorial(N-3))
        #for large N use stirling formula
        #dist[i,1]=np.exp(-b/d)*np.power(d,-N+1)*np.power(b*np.exp(1)/(N-3),N-2)/(np.exp(1)*math.sqrt(2*math.pi/(N-3)))
        #for large N exponentiate the elements
        dist[i,1]=np.exp(-B/d-(N-1)*np.log(d)+(N-2)*np.log(B*np.exp(1.)/(N-3))-1.-0.5*np.log(2*math.pi/(N-3)))
    return dist

print("N Dmax Dav sigma")
dist10=posterior('Data/dataM10d2.5.dat')
dist100=posterior('Data/dataM100d2.5.dat')
dist1000=posterior('Data/dataM1000d2.5.dat')
```

```python
# Plot the Posterior Distribution of the Diffusion coefficient
plt.plot(dist10[:,0],dist10[:,1],color='black',linestyle='-.',label=r'$M=10$')
plt.plot(dist100[:,0],dist100[:,1],color='black',linestyle='--',label=r'$M=100$')
plt.plot(dist1000[:,0],dist1000[:,1],color='black',label=r'$M=1000$')
plt.axvline(x=D1,color='black',linestyle=':')
plt.legend()
plt.xlim(0,4)
plt.xlabel('${\cal D}$ ($\mu$m$^{-2}$s$^{-1}$)')
plt.ylabel('Posterior distribution $p({\cal D}|{x_i,y_i})$')
plt.tight_layout()
plt.savefig('posteriordistD.pdf')
plt.show()
```
